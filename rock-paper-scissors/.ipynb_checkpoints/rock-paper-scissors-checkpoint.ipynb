```python
# Import the choice function of the random module
# https://stackoverflow.com/questions/306400/how-to-randomly-select-an-item-from-a-list

import random
# Assign to a list the 3 possible options: 'stone', 'paper' or 'scissors'.
options = ['stone', 'paper', 'scissors']


# Assign a variable to the maximum number of games: 1, 3, 5, etc ...
max_games = 3

# Assign a variable to the number of games a player must win to win.
# Preferably the value will be based on the number of maximum games


# Define a function that randomly returns one of the 3 options.
# This will correspond to the play of the machine. Totally random.
def turn_machine():
    machine_option = random.choice(options)
    return machine_option

# Define a function that asks your choice: 'stone', 'paper' or 'scissors'
# you should only allow one of the 3 options. This is defensive programming.
# If it is not stone, paper or scissors keep asking until it is.

def turn_player():
    player_option =''
    while player_option != '1' and player_option != '2' and player_option != '3':
        print('Choose one option: \n','STONE----> 1 \n', 'PAPER-----> 2 \n','SCISSORS----> 3 \n')
        player_option = input()
        return player_option

# Define a function that resolves a combat.
# Returns 0 if there is a tie, 1 if the machine wins, 2 if the human player wins
def game(mach_opt, pla_opt):
    game = 0
    if(mach_opt == 'stone' and pla_opt=='1'):
        print('Machine: Stone, Player: Stone')
        game = 0
        return game
    elif(mach_opt == 'stone' and pla_opt=='2'):
        print('Machine: Stone, Player: Paper')
        game = 2
        return game
    elif(mach_opt == 'stone' and pla_opt=='3'):
        print('Machine: Stone, Player: Scissors')
        game = 1
        return game
    elif(mach_opt == 'paper' and pla_opt == '1'):
        print('Machine: Paper, Player: Stone')
        game = 1
        return game
    elif(mach_opt == 'paper' and pla_opt == '2'):
        print('Machine: Paper, Player: Paper')
        game = 0
        return game
    elif(mach_opt == 'paper' and pla_opt == '3'):
        print('Machine: Paper, Player: Scissors')
        game = 2
        return game
    elif(mach_opt == 'scissors' and pla_opt == '1'):
        print('Machine: Scissors, Player: Stone')
        game = 2
        return game
    elif(mach_opt == 'scissors' and pla_opt == '2'):
        print('Machine: Scissors, Player: Paper')
        game = 1
        return game
    elif(mach_opt == 'scissors' and pla_opt == '3'):
        print('Machine: Scissors, Player: Scissors')
        game = 0
        return game
        

# Define a function that shows the choice of each player and the state of the game
# This function should be used every time accumulated points are updated

def state_game():
    result = game(turn_machine(),turn_player())
    if(result == 0):
        print('Tie')
    elif(result == 1):
        print('Machine Wins!')
        return 1
    else:
        print('Player Wins!')
        return 2
    

# Create two variables that accumulate the wins of each participant


# Create a loop that iterates while no player reaches the minimum of wins
# necessary to win. Inside the loop solves the play of the
# machine and ask the player's. Compare them and update the value of the variables
# that accumulate the wins of each participant.
rounds = 0
machine_wins = 0
player_wins = 0
while(rounds < max_games):
    winner = state_game()
    rounds += 1
    if (winner == 1):
        machine_wins += 1
    elif(winner == 2):
        player_wins += 1



# Print by console the winner of the game based on who has more accumulated wins

if(machine_wins > player_wins):
    print('Machine Win the Game!')
elif(machine_wins < player_wins):
    print('Player Win the Game!')
else:
    print('Tie no wins!')
```
